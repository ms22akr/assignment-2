### Correlation Analysis

#### Correlation between the two Energy use and population growth

```python
# calculate the correlation between the two Energy use and population growth
corr = df['Energy use'].corr(df['Population growth'])
print(round(corr,4))
```

A correlation of -0.1802 between Energy use and population growth for all countries from 2000 to 2021 indicates a weak negative correlation between these two variables. This means that as population growth increases, energy use tends to decrease slightly, but the relationship between the two is not very strong. 

#### Correlation between the two Energy use and access to electricity

```python
# calculate the correlation between the two Energy use and access to electricity
corr = df['Energy use'].corr(df['access to electricity'])
print(round(corr,4))
```

The correlation coefficient measures the strength and direction of a linear relationship between two variables. A value of 0.4952 indicates a moderate positive correlation between energy use and access to electricity for all countries from 2000 to 2021. This means that as energy use increases, access to electricity also tends to increase, and vice versa.

#### Correlation between the two CO2 emissions (kt) and electric power consumption

```python
# calculate the correlation between the two CO2 emissions (kt) and electric power consumption
corr = df['CO2 emissions (kt)'].corr(df['EPC'])
print(round(corr,4))
```

The correlation coefficient measures the strength and direction of a linear relationship between two variables. A value of 0.2078 indicates a weak positive correlation between CO2 emissions (measured in kilotons) and electric power consumption. This means that as CO2 emissions increase, electric power consumption also tends to increase, but the relationship is not very strong.

#### Correlation between the two CO2 emissions (kt) and Agricultural land (sq. km)

```python
# Calculate the correlation between the two CO2 emissions (kt) and Agricultural land (sq. km)
corr = df['CO2 emissions (kt)'].corr(df['Agricultural land (sq. km)'])
print(round(corr,4))
```

A correlation coefficient of 0.4964 indicates a moderately positive correlation between the two variables, CO2 emissions (kt) and Agricultural land (sq. km). This means that as the value of one variable increases, the value of the other variable tends to increase as well, but not necessarily at the same rate.
